```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_hindi = pd.read_csv('cleaned_bollywood.csv')
df_telugu = pd.read_csv('cleaned_tolly_movies.csv')
df_indian = pd.read_csv('cleaned_indian_movies.csv')
```

```python

# Made a dataframe wich merges all 3 dfs to find top 3 across all languages
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index = True)
```

```python
avg_genre = round(df_all.groupby('Genre')['BoxOffice (Crores)'].mean(), 2).reset_index()
```

```python
avg_genre
```

```python
avg_genre.sort_values('BoxOffice (Crores)', ascending = False).plot.barh(x= 'Genre', title = 'All Indian movies Genre avg box office')


     
```

# Inferences:

### 1. The top-performing genres in Indian cinema include Epic Fantasy Action, biographical sports dramas, fantasy, action, and science fiction.
### 2. These genres have strong mass appeal and are generally focused on entertainment, which helps them succeed at the box office.
### 3. On the other hand, genres like sports dramas, sports action, thrillers, and family dramas tend to underperform.
### 4. These lower-performing genres often involve deeper or more critical storytelling, which may not attract as wide an audience.


```python
avg_genre_bolly = round(df_hindi[df_hindi['Language'] == 'Hindi'].groupby('Genre')['BoxOffice (Crores)'].mean(), 2).reset_index()
```

```python
avg_genre_bolly.sort_values('BoxOffice (Crores)', ascending = False).plot.barh(x= 'Genre', title = 'Bollywood Genre avg box office')
```

# Inferences:

### 1. Bollywood appears to have fewer genres overall, indicating less variety compared to the broader Indian film industry.
### 2. Biographical sports dramas perform well in Bollywood too, showing consistent audience interest in this genre.
### 3. Genres like comedy, action, and romance seem to attract more attention in Bollywood than in the overall Indian cinema landscape.
### 4. Family dramas and thrillers tend to be the lowest performing genres, possibly because Bollywood gives more focus to historical and biographical storytelling.
