```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_hindi = pd.read_csv('cleaned_bollywood.csv')
df_telugu = pd.read_csv('cleaned_tolly_movies.csv')
df_indian = pd.read_csv('cleaned_indian_movies.csv')
```

```python

# creating a new column called ROI where profit = boxoffice - budget , ROI = profit/budget
df_hindi['ROI'] = (df_hindi['BoxOffice (Crores)'] - df_hindi['Budget (Crores)']) / df_hindi['Budget (Crores)']
df_telugu['ROI'] = (df_telugu['BoxOffice (Crores)'] - df_telugu['Budget (Crores)']) / df_telugu['Budget (Crores)']
df_indian['ROI'] = (df_indian['BoxOffice (Crores)'] - df_indian['Budget (Crores)']) / df_indian['Budget (Crores)']
```

```python
# rounded the figure to two spaces
df_hindi['ROI'] = round(df_hindi['ROI'],2)
df_telugu['ROI'] = round(df_telugu['ROI'],2)
df_indian['ROI'] = round(df_indian['ROI'],2)
```

```python

# Finding top 3 Hindi films(ROI wise) uisng filters of only choosing where language is hindi and then sorting and finally choosing top_3
top_3_hindi = df_hindi.sort_values('ROI', ascending=False).head(3)
top_3_hindi
```

```python
import matplotlib.pyplot as plt

top_3_hindi.plot(x='Title', y='ROI', kind='bar', title='ROI for Top 3 Hindi Films', legend=False)
plt.ylabel('ROI')
plt.xlabel('Movie Title')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Finding top 3 Telugu films(ROI wise) by sorting and finally choosing top_3
top_3_telugu = df_telugu.sort_values('ROI', ascending=False).head(3)
top_3_telugu
```

```python
import matplotlib.pyplot as plt

# Plot the bar graph
ax = top_3_telugu.plot(x='Title', y='ROI', kind='bar', title='ROI for Top 3 Telugu Films', legend=False)

# Add labels for each bar
for i, value in enumerate(top_3_telugu['ROI']):
    ax.text(i, value + 0.05, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

# Set the labels and adjust x-ticks
plt.ylabel('ROI')
plt.xlabel('Movie Title')
plt.xticks(rotation=45)

# Make sure everything fits
plt.tight_layout()
plt.show()
```

```python
# Made a dataframe wich merges all 3 dfs to find top 3 across all languages
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index = True)
```

```python
# Sorting and choosing top 3 of all languages df
top_3_all = df_all.sort_values('ROI', ascending=False).head(3)
top_3_all
```

```python
import matplotlib.pyplot as plt

# Plot the bar graph
ax = top_3_all.plot(x='Title', y='ROI', kind='bar', title='ROI for Top 3 Films', legend=False)

# Add labels for each bar
for i, value in enumerate(top_3_all['ROI']):
    ax.text(i, value + 0.05, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

# Set the labels and adjust x-ticks
plt.ylabel('ROI')
plt.xlabel('Movie Title')
plt.xticks(rotation=45)

# Ensure everything fits
plt.tight_layout()
plt.show()
```
