```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_hindi = pd.read_csv('cleaned_bollywood.csv')
df_telugu = pd.read_csv('cleaned_tolly_movies.csv')
df_indian = pd.read_csv('cleaned_indian_movies.csv')
```

```python

# Made a dataframe wich merges all 3 dfs to find top 3 across all languages
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index = True)
```

```python

df_all['BoxOffice (Crores)'].mean()
```

```python

df_all['BoxOffice (Crores)'].plot(kind='hist', bins=10, edgecolor = 'Black', title='Box Office Distribution')
```

```python
plt.boxplot(df_all['BoxOffice (Crores)'])
plt.title('BoxOffice (Crores)')
plt.ylabel('Revenue')
plt.show()
```

# Inferences from the histogram and boxplot:

### 1. The histogram shows that most movies fall into the box office category of under 300 crores.
### 2. The boxplot confirms this, with the majority of movies earning between 0 and 300 crores.
### 3. The orange line in the boxplot indicates the median box office collection is around 296 crores.
### 4. The small floating circles represent outliers, which are a few movies that earned significantly more than the rest.
