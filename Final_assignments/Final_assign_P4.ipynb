```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_hindi = pd.read_csv('cleaned_bollywood.csv')
df_telugu = pd.read_csv('cleaned_tolly_movies.csv')
df_indian = pd.read_csv('cleaned_indian_movies.csv')
```

```python

# Made a dataframe wich merges all 3 dfs to find top 3 across all languages
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index = True)
```

```python

df_all.info()
```

# Analysis 1

```python

# Plotting for Release year and their avg box office
yearly_boxoffice = df_all.groupby('ReleaseYear')['BoxOffice (Crores)'].mean().plot(kind='line', title ='Release Year / avg box office')
```

# Inferences:

### 1. There was a noticeable increase in movie performance around 2012–2016, likely due to the rise of multiplexes like PVR.
### 2. Growth in the urban middle class also contributed to increased movie consumption during this period.
### 3. Enhanced marketing and promotion through social media and digital platforms further boosted visibility and audience reach.
### 4. The dip seen around 2019–2020 is a result of the COVID-19 pandemic and its widespread impact on the film industry.


# Analysis 2

```python
avg_rating_language = df_all.groupby('Language')['Rating'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(18, 5))

# Plot 1: Avg Rating by Language
avg_rating_language.plot(kind='bar')
plt.title('Average Movie Rating by Language')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
```

# Inferences:

### 1. Movies dubbed from english have high avg rating.
### 2. respective languague films are rated avg good 
### 3. By this we can suggest which type of movies are liked by audiances

# Analysis - 3

```python
top_actors = df_all.groupby('LeadActor')['BoxOffice (Crores)'].mean().sort_values(ascending=False).reset_index()


     
```

```python

top_actors
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
top_actors.sort_values('BoxOffice (Crores)', ascending=False).plot.barh(
    x='LeadActor',
    title='Actors and Avg Collection',
    legend=False
)
plt.tight_layout()
plt.show()
```

# Inferences:

### 1. Movies with famous or superstar lead actors show a clear increase in average box office collections.
### 2. In contrast, films featuring lesser-known actors tend to generate lower revenue.
### 3. This indicates that star power plays a major role in driving audience interest and commercial success.
