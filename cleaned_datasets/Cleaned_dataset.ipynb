## Tollywood_movies

```python
df_tolly = pd.read_excel('tollywood_movies.xlsx', engine='openpyxl')
```

```python
df_tolly
```

```python
# dropping unnamed column because the indexes are wrong and also we already have indexes
df_tolly = df_tolly.drop('Unnamed: 0', axis =1)
```

```python
df_tolly
```

```python
df_tolly.info()
```

### Nan values of language are replaced with telugu since they are telugu movies


```python
df_tolly['Language'] = df_tolly['Language'].fillna('Telugu')

```

### Nan value in ProductionCompany is replaced with mode value


```python
mode_value = df_tolly['ProductionCompany'].mode()[0]
df_tolly['ProductionCompany'] = df_tolly['ProductionCompany'].fillna(mode_value)
```

```python
df_tolly['Rating'] = df_tolly['Rating'].fillna(df_tolly['Rating'].mean())
df_tolly['Duration (minutes)'] = df_tolly['Duration (minutes)'].fillna(df_tolly['Duration (minutes)'].mean())

```

```python
df_tolly['Duration (minutes)'] = df_tolly['Duration (minutes)'].round()
df_tolly['Rating'] = df_tolly['Rating'].round(1)  # one decimal place
```

```python
df_tolly = df_tolly.drop_duplicates(subset='Title', keep='last')
df_tolly.reset_index(drop=True, inplace=True)
```

```python
df_tolly = df_tolly.iloc[1:].reset_index(drop=True)
```

```python
mode_year = df_tolly['ReleaseYear'].mode()[0]
df_tolly['ReleaseYear'] = df_tolly['ReleaseYear'].fillna(mode_year)
```

```python
df_tolly['ReleaseYear'] = df_tolly['ReleaseYear'].astype(int)
```

```python
df_tolly.info()
```

```python
df_tolly.to_csv('cleaned_tolly_movies.csv')
```
